Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Import Dataset

```python
df = pd.read_csv('player_shooting_2023_2024.csv')
```

Analyze the dataset

```python
# Display the first few rows of a DataFrame

df.head()
```

```python
# Summary of a DataFrame

df.info()
```

```python
# Generate descriptive statistics for the numerical columns of a DataFrame

df.describe()
```

```python
# Check for duplicate rows in the DataFrame

df.duplicated().sum()
```

```python
# check the number of missing values (NaN values) in each column of a DataFrame
df.isnull().sum()
```

```python
# Retrieve the column labels (column names) of a DataFrame

df.columns
```

```python
df
```

```python
# Identify and select the column names of categorical features (variables) in a DataFrame

categorical_features = df.select_dtypes(include=['object']).columns.tolist()
```

```python
categorical_features
```

```python
# Fill missing values in categorical features of a DataFrame
for feature in categorical_features:
    most_frequent_value = df[feature].mode()[0]
    df[feature] = df[feature].fillna(most_frequent_value)
```

```python
numerical_features = df.select_dtypes(include=['int', 'float']).columns.tolist()
for feature in numerical_features:
    mean_value = df[feature].mean()
    df[feature] = df[feature].fillna(mean_value)
```

```python

# Re check the number of missing values (NaN values) in each column of a DataFrame
df.isnull().sum()
```

Data is clean now

```python
# Calculating unique values
for i in df.columns:
    print(i,':',df[i].nunique())
```

```python
#  Index object containing the column labels (names) of the DataFrame

df.columns
```

```python
# groupby operation on the DataFrame data based on the 'Player' column and aggregates information for each player

player_info=df.groupby('Player').agg({
    'Nation': 'first',
    'Pos': 'first',
    'Squad': 'first',
    'Age': 'first',
    'Gls': 'sum'  # Total goals scored by the player
}).reset_index()
```

```python
# Top 10 players with the highest number of goals
top_player_info=player_info.sort_values(by='Gls', ascending=False)
top_player_info.head(10)
```

```python
# Value in the 'Player' column is equal to the string 'Player'

df[df['Player']=='Player']
```

```python
# Create a bar plot showing the top 10 players by the number of goals scored.

sns.barplot(data=top_player_info.head(10), x='Player', y='Gls', hue='Pos')
plt.title('Top Players by Goals Scored')
plt.xlabel('Player')
plt.ylabel('Goals Scored')
plt.xticks(rotation=90)
plt.plot()

```

```python
# Bar plot showing the count of players from each nation in the dataset

df['Nation'].value_counts().sort_values(ascending=False).plot(kind='bar', figsize=(10, 6))
plt.title('Count of Players by Nation')
plt.xlabel('Nation')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.plot()
```

```python
# Shooting statistics for each player in the DataFrame . Top players based on the total number of shots taken

player_Sh_info=df.groupby('Player').agg({
    'Nation': 'first',
    'Pos': 'first',
    'Squad': 'first',
    'Age': 'first',
    'Sh': 'sum' # Total shots taken by the player
    ,'FK':'sum',
    'PK':'sum'

}).reset_index()
top_Sh_player_info=player_Sh_info.sort_values(by='Sh', ascending=False)
```

```python
sns.barplot(data=top_Sh_player_info.head(10),x='Player',y='Sh')
plt.title('Top Players by Shots Scored')
plt.xlabel('Player')
plt.ylabel('Shots Scored')
plt.xticks(rotation=90)
plt.plot()
```

```python
# # Number of free kicks scored ('FK' column), selects the top 10 players

top_FK_player_info = player_Sh_info.sort_values(by='FK', ascending=False)
sns.barplot(data=top_FK_player_info.head(10), x='Player', y='FK', hue='Player')
plt.title('Top Players by Free kicks Scored')
plt.xlabel('Player')
plt.ylabel('Free kicks Scored')
plt.xticks(rotation=90)
plt.plot()
```

```python
# Top 10 players by the number of penalty kicks made

top_PK_player_info = player_Sh_info.sort_values(by='PK', ascending=False)
sns.barplot(data=top_PK_player_info.head(10), x='Player', y='PK', hue='Player')
plt.title('Top Players made Penalty kicks ')
plt.xlabel('Player')
plt.ylabel('Penalty kicks')
plt.xticks(rotation=90)
plt.plot()
```

```python
# Histogram to visualize the distribution of ages in a dataset

df['Age'].hist(bins=20, color='pink', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of Expected Goals (xG)

df['xG'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Expected Goals (xG)')
plt.xlabel('xG')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of Non-penalty Expected Goals (npxG).

df['npxG'].hist(bins=20, color='green', edgecolor='black')
plt.title('Distribution of Non-penalty Expected Goals (npxG)')
plt.xlabel('npxG')
plt.ylabel('Frequency')
plt.show()
```

```python
# Relationship between G-xG and np:G-xG

plt.figure(figsize=(8, 6))
plt.scatter(df['G-xG'], df['np:G-xG'], color='skyblue', alpha=0.5)
plt.title('Relationship between G-xG and np:G-xG')
plt.xlabel('G-xG (Difference between actual and expected goals)')
plt.ylabel('np:G-xG (Difference between non-penalty actual and expected goals)')
plt.grid(True)
plt.show()
```

In summary, A scatter plot to visualize the relationship between the difference between actual goals and expected goals and the difference between non-penalty actual goals and expected goals. The plot helps to understand if there is any correlation or pattern between these two variables.

```python
# Relationship between PKatt and PK

plt.scatter(df['PKatt'], df['PK'], color='orange', alpha=0.5)

# Add title and axis labels
plt.title('Relationship between PKatt and PK')
plt.xlabel('PKatt (Penalty kick attempts by the player)')
plt.ylabel('PK (Penalty kicks made by the player)')

# Add grid lines
plt.grid(True)
```

In summary, A scatter plot to visualize the relationship between the number of penalty kick attempts and the number of penalty kicks made by players. The plot helps to understand if there is any correlation or pattern between these two variables.
